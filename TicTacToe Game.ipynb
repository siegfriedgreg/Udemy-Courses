```python
'''
Author: Greg Siegfried
Date: 8/2/2021
Purpose: Milestone 1 Assignment from Udemy Course
Project: Two Player Tic-Tac-Toe game
Operation: Player is given the option to play or play again.
            Player One is given the choice of game piece type X or O
            Player One chooses first move from game board:
                     1 | 2 | 3 
                    -----------
                     4 | 5 | 6 
                    -----------
                     7 | 8 | 9 
            If player one has not won yet, player two will get the next turn.
            Game is over when either player get's three of their pieces in a row, column, or diagonal

'''
from IPython.display import clear_output

# GLOBAL VARIABLES
game_state = True
game_on = True
game_piece = ['',"X","O"]
game_board = ['','1','2','3','4','5','6','7','8','9']
player_one = ''
player_two = ''

# GLOBAL FUNCTIONS
def print_gameboard():
    # PRINTS THE CURRENT STATE OF THE GAME BOARD
    row1 = " {} | {} | {} ".format(game_board[1], game_board[2], game_board[3])
    row2 = " {} | {} | {} ".format(game_board[4], game_board[5], game_board[6]) 
    row3 = " {} | {} | {} ".format(game_board[7], game_board[8], game_board[9]) 
    div = "-----------"
    
    print(row1)
    print(div)
    print(row2)
    print(div)
    print(row3)

def player_choice(player_id):
    # PLAYER MAKES A CHOICE TILL ACCEPTABLE ANSWER
    valid = False
    while not valid:
        choice = int(input("Make a move by choosing an available square: 1-9 \n"))
        if choice in range(1,10) and game_board[choice] == str(choice):
            valid = True
            game_board[choice] = player_id
        else:
            print("Not A Valid Move! \n")
            
def check_win_state(p_id):
    
    # ROW CHECK
    if game_board[1] == p_id and game_board[2] == p_id and game_board[3] == p_id:
        game_on = False 
        return True
    elif game_board[4] == p_id and game_board[5] == p_id and game_board[6] == p_id:
        game_on = False
        return True
    elif game_board[7] == p_id and game_board[8] == p_id and game_board[9] == p_id:
        game_on = False
        return True
        
    # COLUMN CHECK
    elif game_board[1] == p_id and game_board[4] == p_id and game_board[7] == p_id:
        game_on = False
        return True
    elif game_board[2] == p_id and game_board[5] == p_id and game_board[8] == p_id:
        game_on = False
        return True
    elif game_board[3] == p_id and game_board[6] == p_id and game_board[9] == p_id:
        game_on = False
        return True
        
    # DIAGONAL CHECK
    elif game_board[1] == p_id and game_board[5] == p_id and game_board[9] == p_id:
        game_on = False
        return True
    elif game_board[3] == p_id and game_board[5] == p_id and game_board[7] == p_id:
        game_on = False
        return True
    else:
        return False

# GAME START ------------------------------------------------
while game_state:
    
    # PLAY / REPLAY
    replay = str(input("Do you want to play a game? y/n \n"))
    if replay == 'n':
        game_state = False
        clear_output()
        break
        
    # CLEAR GAME BOARD
    clear_output()
    game_board = ['','1','2','3','4','5','6','7','8','9']
    
    # PLAYER PIECE CHOICE
    print("Player One, please choose a game piece:")
    piece_choice = int(input(" 1 = 'X'   2 = 'O'\n"))
    if piece_choice == 1:
        player_one = game_piece[1]
        player_two = game_piece[2]
    elif piece_choice == 2:
        player_one = game_piece[2]
        player_two = game_piece[1]
    else:
        print("Incorrect choice, try again")
        break
    
    # CLEAR OUTPUT
    clear_output()
    
    # PRINT GAME BOARD
    print_gameboard()
    
    while game_on:        

        # PLAYER_ONE SELECT MOVE
        player_choice(player_one)
        if check_win_state(player_one):
            clear_output()
            print(" Congratulations Player One, you win! \n")
            print_gameboard()
            break
        
        clear_output()
        print_gameboard()
            
        # PLAYER_TWO SELECT MOVE
        player_choice(player_two)
        if check_win_state(player_two):
            clear_output()
            print(" Congratulations Player Two, you win! \n")
            print_gameboard()
            break
        
        clear_output()
        print_gameboard()

    # REPEAT GAME_ON WHILE LOOP TILL WINNER

# REPEAT GAME_STATE WHILE LOOP TILL REPLAY CHOICE == NO
     
# GAME END --------------------------------------------------        
```
